# Pre_Processing

Preprocessing data adalah tahap awal dalam analisis data yang bertujuan untuk membersihkan dan mempersiapkan data agar dapat digunakan dengan efektif oleh model atau alat analisis data. Proses ini melibatkan serangkaian langkah atau teknik untuk mengatasi masalah dan kelemahan dalam dataset sehingga model atau alat analisis data dapat memberikan hasil yang lebih akurat dan bermakna.

Langkah-langkah Pre-Processing:
1. Punctuation process
2. Stopword
3. Tokenisasi
4. TF-IDF

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

import warnings
import pandas as pd
import numpy as np
import nltk
import re
import csv

nltk.download('stopwords')
nltk.download('punkt')
warnings.filterwarnings('ignore')
```

##Punctuation process
Punctuation process adalah salah satu tahap dalam pre-processing data yang melibatkan penghapusan tanda baca dari teks atau dataset. Tanda baca seperti titik, koma, tanda tanya, dan lainnya dianggap tidak memberikan informasi penting.

```python
def cleaning(text):
  text = re.sub(r'[^a-zA-Z\s]', '', text).strip()
  return text

df['data_clean'] = df['Abstrak'].apply(cleaning)
df['data_clean']
```

```python
def cek_specialCharacter(dokumen):
  karakter = ['!', '@', '#', '$', '%', '^', '&', '*', '(', ')', '-', '_', '+', '=', '{', '}', '[', ']', '|', '\\', ':', ';', '"', "'", '<', '>', ',', '.', '?', '/', '`', '~']
  for i in dokumen:
    if i in karakter :
      print(dokumen)
df['data_clean'].apply(cek_specialCharacter)
```

##tokenizing
Tokenizing adalah proses pemisahan teks atau dokumen menjadi unit-unit diskrit yang lebih kecil yang disebut sebagai token.

```python
def tokenizer(text):
  text = text.lower()
  return word_tokenize(text)

df['Tokenizing'] = df['data_clean'].apply(tokenizer)
df['Tokenizing']
```

##stopword
Stopword adalah kata-kata umum yang sering muncul dalam suatu teks dan dianggap memiliki sedikit nilai informatif atau kontribusi terhadap pemahaman makna. seperti ke,di,dan,atau dll.

```python
corpus = stopwords.words('indonesian')

def stopwordText(words):
 return [word for word in words if word not in corpus]

df['Stopword Removal'] = df['Tokenizing'].apply(stopwordText)

# Gabungkan kembali token menjadi kalimat utuh
df['stopword'] = df['Stopword Removal'].apply(lambda x: ' '.join(x))
df['stopword']
```

##tf-idf
Term Frequency-Inverse Document Frequency (TF-IDF) adalah metode statistik yang digunakan untuk mengevaluasi seberapa penting suatu kata dalam suatu dokumen dalam konteks koleksi dokumen. biasanya digunakan untuk memberikan bobot pada setiap kata-kata dalam dokumen.

```python
def tfidf(dokumen):
  vectorizer = TfidfVectorizer()
  x = vectorizer.fit_transform(dokumen).toarray()
  terms = vectorizer.get_feature_names_out()

  final_tfidf = pd.DataFrame(x, columns=terms)
  final_tfidf.insert(0, 'Abstrak', dokumen)

  return (vectorizer, final_tfidf)

tfidf_vectorizer, final_tfidf = tfidf(df['stopword'])
final_tfidf
```

```python
final_tfidf.to_csv('HasilTF-IDF-new.csv', index=False)
```

##term frequensi
Term Frequency (TF) adalah ukuran seberapa sering suatu kata atau term muncul dalam suatu dokumen. Tujuan dari menghitung term frequency adalah untuk memberikan bobot pada kata-kata dalam suatu dokumen sehingga kita dapat memahami seberapa penting atau relevan kata tersebut terhadap dokumen tersebut.

```python
def term_freq(dokumens):
  # Buat objek CountVectorizer
  vectorizer = CountVectorizer()
  tf_matrix = vectorizer.fit_transform(dokumens).toarray()
  terms = vectorizer.get_feature_names_out()

  final_tf = pd.DataFrame(tf_matrix, columns=terms)
  final_tf.insert(0, 'Abstrak', dokumens)

  return (vectorizer, final_tf, tf_matrix, terms)

tf_vectorizer, final_tf, tf_matrix, tf_terms = term_freq(df['stopword'])
final_tf
```

```python
final_tf.to_csv('HasilTermFrequensi-new.csv', index=False)
```

##logarithm frequensi
Log-frequency atau log-frequency weighting adalah pendekatan yang digunakan dalam analisis teks atau pemrosesan bahasa alami untuk mengukur seberapa sering suatu kata muncul dalam suatu dokumen atau korpus dengan menggunakan skala logaritmik.

```python
def logarithm_freq(dokumens):
  return np.log10(dokumens + 1)

df_logarithm_freq = pd.DataFrame(tf_matrix, columns=tf_terms).apply(logarithm_freq)
df_logarithm_freq.insert(0, 'Abstrak', df['stopword'])
df_logarithm_freq
```
