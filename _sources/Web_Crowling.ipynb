# Web_Crowling


####Penjelasan
Pengertian Crawling?
Crawling adalah proses yang dilakukan oleh bot atau web crawler untuk mengumpulkan informasi dari berbagai situs web secara otomatis. Aktivitas ini merupakan bagian dari proses web scraping atau pengambilan informasi dari halaman web.
Tujuan Crawling?
1. Pengindeksan Mesin Pencari: Salah satu tujuan utama dari crawling adalah memungkinkan mesin pencari untuk mengumpulkan dan mengindeks informasi dari berbagai situs web di seluruh internet. Mesin pencari seperti Google menggunakan web crawler untuk menjelajahi halaman-halaman web dan mengumpulkan informasi yang kemudian diindeks, sehingga pengguna dapat menemukan konten yang relevan saat melakukan pencarian.

2. Pemantauan Perubahan: Crawling juga dapat digunakan untuk memantau perubahan pada situs web tertentu. Misalnya, perusahaan atau individu dapat menggunakan web crawler untuk secara otomatis memeriksa perubahan pada halaman-halaman web, mengumpulkan data terbaru, dan memberikan notifikasi jika terjadi perubahan yang signifikan.

3. Penambangan Data: Crawling merupakan langkah awal dalam proses penambangan data di web (web data mining). Data yang dikumpulkan dari situs web dapat dianalisis lebih lanjut untuk mendapatkan wawasan atau informasi yang berguna. Ini dapat melibatkan ekstraksi informasi tertentu, analisis sentimen, atau identifikasi tren.

4. Analisis dan Riset Pasar: Crawling dapat digunakan dalam analisis pasar dan riset untuk mengumpulkan data tentang produk, harga, ulasan pelanggan, dan informasi lainnya dari situs-situs web e-commerce atau sumber informasi lainnya. Ini membantu perusahaan dalam membuat keputusan strategis.

5. Pengumpulan Berita atau Konten: Beberapa organisasi menggunakan crawling untuk mengumpulkan dan menyaring berita atau konten tertentu dari berbagai sumber. Ini dapat membantu dalam membuat agregator berita atau menyediakan konten yang relevan untuk audiens tertentu.

6. Monitoring Kinerja dan SEO: Web crawler dapat digunakan untuk memantau kinerja suatu situs web, seperti waktu respon dan status halaman. Selain itu, crawler dapat membantu dalam memeriksa faktor-faktor SEO (Search Engine Optimization), seperti struktur tautan dan kata kunci, untuk membantu situs web mendapatkan peringkat yang lebih baik di mesin pencari.

7. Pemantauan Kompetitor: Beberapa organisasi menggunakan crawling untuk memantau kegiatan pesaing mereka, mengumpulkan informasi tentang produk, harga, promosi, dan strategi lainnya.



Beberapa library yang dibutuhkan untuk scrapping data dari situs web.

```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings("ignore")
headers = {
        'Access-Control-Allow-Origin': '*',
        'Access-Control-Allow-Methods': 'GET',
        'Access-Control-Allow-Headers': 'Content-Type',
        'accept': '/',
        'accept-encoding': 'gzip, deflate',
        'accept-language': 'en,mr;q=0.9',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36'}
```

Script untuk melakukan scrapping data dari website PTA Trunojoyo (judul,penulis,pembimbing 1,pembimbing 2, dan abstrak)

```python
list_judul = []
list_penulis = []
list_pembimbing1 = []
list_pembimbing2 = []
list_abstrak = []
url = f"https://pta.trunojoyo.ac.id/c_search/byprod/28/"
req = requests.get(url,headers=headers,timeout=5,verify=False)
for i in range(2, 71):
    soup = BeautifulSoup(req.content, 'html.parser')
    results = soup.find(id="content_journal")
    elements = results.find_all("li", attrs={"data-id":"id-1"})
#     print(elements)
    for element in elements:
        info = element.find_all("div")[0]
        info_nama = info.find_all("div")
#         print(info_nama)
        link_abstrak = element.find_all("div")[-1].find("a")["href"]
#         ekstrak info
        judul = info.find("a", class_="title").text
        penulis, pembimbing1, pembimbing2 = [nama.text.split(":")[1].strip() for nama in info_nama]

#         ekstrak abstrak
#         print(link_abstrak)
        req = requests.get(link_abstrak,headers=headers,timeout=5,verify=False)
        soup_abstrak =  BeautifulSoup(req.content, 'html.parser')
        results_abstrak = soup_abstrak.find_all("p")

        abstrak = results_abstrak[0].text

        list_judul.append(judul)
        list_penulis.append(penulis)
        list_pembimbing1.append(pembimbing1)
        list_pembimbing2.append(pembimbing2)
        list_abstrak.append(abstrak)

    url = f"https://pta.trunojoyo.ac.id/c_search/byprod/28/{i}"
    req = requests.get(url,headers=headers,timeout=50,verify=False)
    if(req.status_code == 404):
        break
```

Menyimpan hasil scrapping sebagai DataFrame

```python
df = pd.DataFrame({
    "Judul" : list_judul,
    "Nama Penulis" : list_penulis,
    "Pembimbing I" : list_pembimbing1,
    "Pembimbing II" : list_pembimbing2,
    "Abstrak" : list_abstrak
})
df
```

Menyimpan hasil dalam bentuk file csv

```python
df.to_csv("dataPTA.csv", index=True)
```
